# Импорты

```
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
import nltk
nltk.download('punkt')
from nltk import sent_tokenize
import re
import gensim
from gensim.models import Word2Vec, doc2vec
```

# Предобработка

```
text = open('anna_karenina.txt').read().lower()[:10000]
len(text)
```

```
sents = sent_tokenize(text)
```

```
toks = []
for line in sents:
    line_tokens = re.findall('\w+', line) 
    toks.append(line_tokens)
```

```
print(toks)
```

```
lems = []
for line in toks:
    line_lems = []
    for tok in line:
        line_lems.append(morph.parse(tok)[0].normal_form)
    lems.append(line_lems)
```

```
print(lems)
```

# Обучение

```
# на вход подаём список списков токенов
# size -- размерность вектора
# min_count -- сколько раз должно встречаться слово (иначе игнорим)
# window -- сколько слов справа и слева считаем контекстом
ru_w2v = Word2Vec(lems, size=10, min_count=1, window=5) 
```

```
print(ru_w2v.wv.vocab) # словарь модели
```

```
ru_w2v.wv.get_vector('анна') # получить у модели вектор для слова
```

```
ru_w2v.wv.get_vector('анна') # получить вектор слова
```

```
ru_w2v.wv.most_similar('анна') # найти слова, ближайшие к данному 
```

```
ru_w2v.wv.similarity('анна', 'каренин') # посчитать близость между словами
```
